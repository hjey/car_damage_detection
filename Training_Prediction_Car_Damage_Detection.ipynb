```python
from ultralytics import YOLO

def train_yolo_model(yaml_path, epochs=500, batch=16, optimizer='Adam'):
    """YOLOv8 모델 학습"""
    # GPU 사용 가능 여부 확인
    import torch
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    print(f"Using device: {device}")

    # 모델명 생성
    model_name = f'yolov8_car_damage_epoch{epochs}_batch{batch}_{optimizer}'

    model = YOLO('yolov8n.yaml')  # 'n'은 nano 버전

    results = model.train(
        data=yaml_path,
        epochs=epochs,
        batch=batch,
        optimizer=optimizer,
        name=model_name,
        device=[0],
        #patience=50,
        save_period=10,
    )

    return model, results
```

```python
# 모델 학습 실행
print("\n모델 학습 시작...")
os.environ['CUDA_LAUNCH_BLOCKING'] = "1"  # 학습 호출 전에 추가
model, results, model_name = train_yolo_model(yaml_path)
print("모델 학습 완료")
```

# 예측 및 시각화

```python
# 8. 테스트 이미지로 예측 및 시각화
def visualize_predictions(model, image_paths, num_samples=9, class_names=None):
    """여러 테스트 이미지의 예측 결과 시각화"""
    import random

    # 클래스 이름 설정
    if class_names is None:
        class_names = {
            0: 'Scratch',
            1: 'Dent',
            2: 'Crash'
        }

    # 이미지가 num_samples보다 적으면 전체 이미지 사용
    n_images = min(len(image_paths), num_samples)

    # 랜덤하게 이미지 선택
    selected_images = random.sample(image_paths, n_images)

    # 서브플롯 그리드 계산
    grid_size = int(np.ceil(np.sqrt(n_images)))
    fig = plt.figure(figsize=(15, 15))

    for idx, image_path in enumerate(selected_images, 1):
        # 예측 수행
        results = model.predict(source=image_path)

        # 서브플롯 추가
        plt.subplot(grid_size, grid_size, idx)
        plt.imshow(results[0].plot())
        plt.axis('off')

        # 파일명과 예측된 클래스들을 제목으로 표시
        file_name = os.path.basename(image_path)
        pred_classes = []
        for r in results[0].boxes:
            cls_id = int(r.cls)
            conf = float(r.conf)
            pred_classes.append(f"{class_names[cls_id]}: {conf:.2f}")

        title = f'{file_name}\n{", ".join(pred_classes)}'
        plt.title(title, fontsize=8)

    plt.tight_layout()
    plt.show()
```

```python
# 테스트 이미지로 예측
test_images = sorted(glob(f'{root_dir}/test_del/images/*.jpg'))
if test_images:
    print("\n테스트 이미지로 예측 실행")
    visualize_predictions(model, test_images, num_samples=9)
else:
    print("No test images found")
```
