```python
from ultralytics import YOLO

def train_yolo_model(yaml_path, epochs=500, batch=16, optimizer='Adam'):
    """YOLOv8 Î™®Îç∏ ÌïôÏäµ"""
    # GPU ÏÇ¨Ïö© Í∞ÄÎä• Ïó¨Î∂Ä ÌôïÏù∏
    import torch
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    print(f"Using device: {device}")

    # Î™®Îç∏Î™Ö ÏÉùÏÑ±
    model_name = f'yolov8_car_damage_epoch{epochs}_batch{batch}_{optimizer}'

    model = YOLO('yolov8n.yaml')  # 'n'ÏùÄ nano Î≤ÑÏ†Ñ

    results = model.train(
        data=yaml_path,
        epochs=epochs,
        batch=batch,
        optimizer=optimizer,
        name=model_name,
        device=[0],
        #patience=50,
        save_period=10,
    )

    return model, results
```

```python
# Î™®Îç∏ ÌïôÏäµ Ïã§Ìñâ
print("\nÎ™®Îç∏ ÌïôÏäµ ÏãúÏûë...")
os.environ['CUDA_LAUNCH_BLOCKING'] = "1"  # ÌïôÏäµ Ìò∏Ï∂ú Ï†ÑÏóê Ï∂îÍ∞Ä
model, results, model_name = train_yolo_model(yaml_path)
print("Î™®Îç∏ ÌïôÏäµ ÏôÑÎ£å")
```

```python
# 8. ÌÖåÏä§Ìä∏ Ïù¥ÎØ∏ÏßÄÎ°ú ÏòàÏ∏° Î∞è ÏãúÍ∞ÅÌôî
def visualize_predictions(model, image_paths, num_samples=9, class_names=None):
    """Ïó¨Îü¨ ÌÖåÏä§Ìä∏ Ïù¥ÎØ∏ÏßÄÏùò ÏòàÏ∏° Í≤∞Í≥º ÏãúÍ∞ÅÌôî"""
    import random

    # ÌÅ¥ÎûòÏä§ Ïù¥Î¶Ñ ÏÑ§Ï†ï
    if class_names is None:
        class_names = {
            0: 'Scratch',
            1: 'Dent',
            2: 'Crash'
        }

    # Ïù¥ÎØ∏ÏßÄÍ∞Ä num_samplesÎ≥¥Îã§ Ï†ÅÏúºÎ©¥ Ï†ÑÏ≤¥ Ïù¥ÎØ∏ÏßÄ ÏÇ¨Ïö©
    n_images = min(len(image_paths), num_samples)

    # ÎûúÎç§ÌïòÍ≤å Ïù¥ÎØ∏ÏßÄ ÏÑ†ÌÉù
    selected_images = random.sample(image_paths, n_images)

    # ÏÑúÎ∏åÌîåÎ°Ø Í∑∏Î¶¨Îìú Í≥ÑÏÇ∞
    grid_size = int(np.ceil(np.sqrt(n_images)))
    fig = plt.figure(figsize=(15, 15))

    for idx, image_path in enumerate(selected_images, 1):
        # ÏòàÏ∏° ÏàòÌñâ
        results = model.predict(source=image_path)

        # ÏÑúÎ∏åÌîåÎ°Ø Ï∂îÍ∞Ä
        plt.subplot(grid_size, grid_size, idx)
        plt.imshow(results[0].plot())
        plt.axis('off')

        # ÌååÏùºÎ™ÖÍ≥º ÏòàÏ∏°Îêú ÌÅ¥ÎûòÏä§Îì§ÏùÑ Ï†úÎ™©ÏúºÎ°ú ÌëúÏãú
        file_name = os.path.basename(image_path)
        pred_classes = []
        for r in results[0].boxes:
            cls_id = int(r.cls)
            conf = float(r.conf)
            pred_classes.append(f"{class_names[cls_id]}: {conf:.2f}")

        title = f'{file_name}\n{", ".join(pred_classes)}'
        plt.title(title, fontsize=8)

    plt.tight_layout()
    plt.show()
```

```python
# ÌÖåÏä§Ìä∏ Ïù¥ÎØ∏ÏßÄÎ°ú ÏòàÏ∏°
test_images = sorted(glob(f'{root_dir}/test_del/images/*.jpg'))
if test_images:
    print("\nÌÖåÏä§Ìä∏ Ïù¥ÎØ∏ÏßÄÎ°ú ÏòàÏ∏° Ïã§Ìñâ")
    visualize_predictions(model, test_images, num_samples=9)
else:
    print("No test images found")
```
